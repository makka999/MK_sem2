```python
# SVD i korelacja obrazu  Wariant 8

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# 1. Wczytanie obrazu
img = Image.open('13.webp').convert('L')  # konwersja do skali szaroci
X = np.array(img)
plt.figure()
plt.title("Obraz oryginalny")
plt.imshow(X, cmap='gray')
plt.axis('off')
plt.show()

# 2. Rozkad SVD
U, S, VT = np.linalg.svd(X, full_matrices=False)
 = np.diag(S)

# 3. Macierze korelacji
XTX = X.T @ X      # korelacja kolumn
XXT = X @ X.T      # korelacja wierszy

# 4. Wizualizacja macierzy korelacji
plt.figure()
plt.title("Macierz korelacji kolumn (XTX)")
plt.imshow(XTX, cmap='hot')
plt.colorbar()
plt.show()

plt.figure()
plt.title("Macierz korelacji wierszy (XXT)")
plt.imshow(XXT, cmap='hot')
plt.colorbar()
plt.show()

# 5. Widok wartoci osobliwych (singular values)
plt.figure()
plt.title("Wartoci osobliwe (Singular Values)")
plt.semilogy(S)
plt.xlabel('Indeks')
plt.ylabel('Warto')
plt.grid(True)
plt.show()
```
