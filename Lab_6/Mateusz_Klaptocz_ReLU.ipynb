```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)

def ReLU(x):
    return np.maximum(0, x)

def dReLU(x):
    return np.where(x > 0, 1, 0)

y = ReLU(x)
dy = dReLU(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='ReLU(x)', color='blue')
plt.plot(x, dy, label="ReLU'(x) — pochodna", color='red', linestyle='--')
plt.title('Funkcja ReLU i jej pochodna')
plt.xlabel('x')
plt.ylabel('Wartość')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
